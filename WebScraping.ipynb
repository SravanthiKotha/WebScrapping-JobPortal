### Code for Web Scrapping (Correct way)

```python
# Scrapping the Web Page

titles = []
ratings = []
prices = []
features = []

for i in range(1, 56):
    URL = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'. format(i)
    
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for container in soup.find_all('div', attrs={'class' : '_2kHMtA'}):

        product = container.find('div', attrs={'class' : '_4rR01T'})
        if product is None:
            titles.append(np.NaN)
        else:
            titles.append(product.text)

        price = container.find('div', attrs={'class' : '_30jeq3 _1_WHN1'})
        if price is None:
            prices.append(np.NaN)
        else:
            prices.append(price.text)

        rating = container.find('div', attrs={'class' : '_3LWZlK'})
        if rating is None:
            ratings.append(np.NaN)
        else:
            ratings.append(rating.text)

        feature = container.find('ul', attrs={'class' : '_1xgFaf'})
        if feature is None:
            features.append(np.NaN)
        else:
            features.append(feature.text)
```

```python
print(len(titles))
print(len(prices))
print(len(ratings))
print(len(features))
```

# Extracting the Details using Regex

```python
import numpy as np
import pandas as pd

import re
```

```python
df = pd.read_csv('Laptop_Details.csv')

df.head()
```

```python
print(df['Feature'][0])
print(df['Feature'][2])
print(df['Feature'][6])
```

```python
# Processor Information

regex = r'^(?:AMD|Intel|M1)[\s\w]+Processor'

df['Feature'].apply(lambda x : re.findall(regex, x))
```

```python
df['Processor'] = df['Feature'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
# RAM Information

regex = r'\d+\sGB[\s\w]+RAM'

df['Feature'].apply(lambda x : re.findall(regex, x))
```

```python
df['RAM'] = df['Feature'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
# OS 

regex = '[\d]+[\s]bit[\s\w]+Operating System'

df['OS'] = df['Feature'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
# Storage

regex = r'[\d]+\s(?:GB|TB)\s(?:HDD|SSD)'

df['Storage'] = df['Feature'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
df['Product'][0]
```

```python
# Brand

regex = r'^\w+'

df['Brand'] = df['Product'].apply(lambda x : re.findall(regex, x))
```
